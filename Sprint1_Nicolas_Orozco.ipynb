```python
import pandas as pd
```

```python
datos = pd.read_csv("Datos.csv")
```

```python
#Aqui se organizan los valores de peso de manera ascendente
datos.sort_values(by='Peso',ascending=True)
```

```python
#Aqui se organizan los valores de peso de manera descendente
datos.sort_values(by='Peso',ascending=False)
```

```python
#Aqui se organizan los valores de edad de manera ascendente
datos.sort_values(by='Edad',ascending=True)
```

```python
#Aqui se organizan los valores de edad de manera descendente
datos.sort_values(by='Edad',ascending=False)
```

```python
datosPesoTrue = datos.sort_values(by='Peso',ascending=True)
datosPesoFalse = datos.sort_values(by='Peso',ascending=False)
```

```python
print("PESO ASCENDENTE")
print(datosPesoTrue)

print("PESO DESCENDENTE")
print(datosPesoFalse)
```

```python
datosEdadTrue = datos.sort_values(by='Edad',ascending=True)
datosEdadFalse = datos.sort_values(by='Edad',ascending=False)
```

```python
print("EDAD ASCENDENTE")
print(datosEdadTrue)

print("EDAD DESCENDENTE")
print(datosEdadFalse)
```
